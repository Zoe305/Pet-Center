```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('공원1.csv', encoding = 'cp949') #데이터를 저장소에 업로드 후 데이터 불러오기
```

```python
df
```

```python
df.columns #데이터프레임의 열 이름 정보 확인하기
```

```python
df1 = df.drop(df.iloc[:, 4:8], axis=1) #필요없는 열 삭제
```

```python
df2 = df1.drop(df1.iloc[:, 7:16], axis=1) #필요한 열 선택
```

```python
df3 = df2.drop(df2.iloc[:, :2], axis=1) #필요한 열 선택
```

```python
df4 = df3.drop('데이터 기준일자', axis=1) #필요없는 열 삭제
```

```python
df5 = df4.drop('이미지명', axis=1) #필요없는 열 삭제
```

```python
df5
```

```python
df5.to_csv('양천구_공원.csv', index = False) #csv 파일로 저장
```

```python
df6 = df5[df5['공원면적']<=100000] #공원 면적이 10만 이하인 공원 추출
```

```python
df6
```

```python
df6.to_csv('양천구_공원_면적10만이하.csv', index = False) #csv 파일로 저장
```

```python
df8 = df5[df5['공원면적']>100000] #공원 면적이 10만제곱 초과인 공원 추출
```
